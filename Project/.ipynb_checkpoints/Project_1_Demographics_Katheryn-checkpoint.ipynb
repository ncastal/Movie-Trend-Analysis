```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress
```

```python
# Read cleaned movie data into pandas dataframe
IMDB_movies_df=pd.read_csv("Resources/IMDB_movies_and_ratings.csv")

# Data clean: remove old index from csv file
del IMDB_movies_df["Unnamed: 0"]

# Print a sample view of the data
IMDB_movies_df.head()
```

```python
# Print a list of columns to get some bearings
print ("Column titles:  ")
print (" ")
for col in IMDB_movies_df.columns:
    print(col)
    
print (" ")
print ("-----------------------")    
print (" ")

# Print the records count to determine level of magnitude
print ("Number of movies in this dataset:  ")
len(IMDB_movies_df)    
```

```python
movies_after_1997=IMDB_movies_df.loc[IMDB_movies_df['year']>=1997].loc[IMDB_movies_df["year"] <= 2017]
movies_after_1997
```

```python
USA_movies=movies_after_1997.loc[movies_after_1997['country']=='USA']
USA_movies=USA_movies.reset_index()
USA_movies.to_csv("Resources/movies_after_1997.csv")
# USA_movies_by_year.to_csv("Resources/USA_movies_by_year_from_1997.csv")
```

```python
# NEW STUFF - KATHERYN
USA_movies_new=USA_movies.drop(['original_title', 'date_published', 'director','language', 'avg_vote','worlwide_gross_income','production_company', 'actors', 'description','duration', 'non_us_voters_rating','non_us_voters_votes','top1000_voters_rating','top1000_voters_votes', 'writer', 'imdb_title_id','us_voters_votes'], axis=1)
USA_movies_new
```

```python
# Create, save to CSV and View dataset on number of viewers by year 
USA_movies_by_year = USA_movies_new.groupby('year').count()
#save data
# USA_movies_by_year.to_csv("Resources/USA_movies_by_year_from_1997.csv")
# show the data
USA_movies_by_year
```

```python
# Reset index and graph plot
USA_movies_by_year.reset_index().plot(x="year", y="title")
# Labels
plt.xlabel("Year")
plt.ylabel("Number of Movies Produced")
plt.title("U.S. Movie Production between 1997-2017")
# Removed legend
plt.legend().set_visible(False)
# Changed x-axis ticks
plt.xticks(range(1997,2018),rotation=90)
plt.show()
```

```python
# SHOWS THE MOST POPULAR GENRES THAT ARE PRODUCED
grouped_genre_count=USA_movies_new.groupby('genre').count()
grouped_genre_count
sort_genre_count=grouped_genre_count.sort_values('title',ascending=False)
sort_genre_count
top_5_genre_produced=sort_genre_count.iloc[0:5]
top_5_genre_produced
```

```python
# PLOT OF THE MOST POPULAR GENRES THAT HAVE BEEN PRODUCED
top_5_genre_produced.plot(kind='bar',color="orange",y=['title'])
# Labels
plt.xlabel("Movie Genre")
plt.ylabel("Number of Movies Produced")
plt.title("Top 5 Movie Genres Produced in the U.S. (1997-2017)")
# Removed legend
plt.legend().set_visible(False)
```

```python
# DON'T INCLUDE
# GENRE PRODUCED THE MOST EACH YEAR - NEED TO FIX
grouped_genre_count2=USA_movies_new.groupby(['year','genre']).count()
grouped_genre_count2
sort_genre_count2=grouped_genre_count2.sort_values('title',ascending=False)
sort_genre_count2
top_5_count2_all=sort_genre_count2.iloc[0:10]
top_5_count2_all
```

```python
# TOP 5 MOST POPULAR GENRES OF MOVIES (by us_voters_rating)
grouped_genre_gen=USA_movies_new.groupby('genre').mean()
grouped_genre_gen
sort_genre=grouped_genre_gen.sort_values('us_voters_rating',ascending=False)
sort_genre
top_5_genre_all=sort_genre.iloc[0:5]
top_5_genre_all
```

```python
#DON'T INCLUDE JUST FOR ME TO VISUALLY SEE IT
# PIVOT TABLE FOR TOP 5 MOST POPULAR GENRES OF MOVIES (by us_voters_rating)
pd.pivot_table(top_5_genre_all, values = 'us_voters_rating',
                          columns='genre')
```

```python
# PLOT OF TOP 5 MOST POPULAR GENRES 
top_5_genre_all.plot(kind='bar',color="green",y=['us_voters_rating'])
# Labels
plt.xlabel("Movie Genre")
plt.ylabel("US Voters Ratings")
plt.title("Top 5 Most Popular Movie Genres According to U.S. Viewers (1997-2017)")
# Removed legend
plt.legend().set_visible(False)
```

```python
# DON'T INCLUDE
# TOP 5 MOST POPULAR GENRES OF MOVIES (by metascore)
# METASCORE is a weighted average of the published critic reviews
sort_genre3=grouped_genre_gen.sort_values('metascore',ascending=False)
sort_genre3
top_5_genre_all3=sort_genre3.iloc[0:5]
top_5_genre_all3
```

```python
# TOP 5 MOST POPULAR GENRES OF MOVIES AMONG MALES
sort_genre_mal=grouped_genre_gen.sort_values('males_allages_avg_vote',ascending=False)
sort_genre_mal
top_5_genre_mal=sort_genre_mal.iloc[0:5]
top_5_genre_mal
```

```python
# DON'T INCLUDE
# PIVOT TABLE FOR TOP 5 MOST POPULAR GENRES OF MOVIES AMOUNG MALES 
pd.pivot_table(top_5_genre_mal, values = 'males_allages_avg_vote',
                          columns='genre')
```

```python
# PLOT OF TOP 5 MOST POPULAR GENRES AMOUNG MALES
top_5_genre_mal.plot(kind='bar',y=['males_allages_avg_vote'])
# Labels
plt.xlabel("Movie Genres")
plt.ylabel("Male Voters Average Ratings")
plt.title("Top 5 Most Popular Movie Genres According to U.S. Male Viewers (1997-2017)")
# Removed legend
plt.legend().set_visible(False)
```

```python
# TOP 5 MOST POPULAR GENRES OF MOVIES AMOUNG FEMALES
sort_genre_fem=grouped_genre_gen.sort_values('females_allages_avg_vote',ascending=False)
sort_genre_fem
top_5_genre_fem=sort_genre_fem.iloc[0:5]
top_5_genre_fem
# top_5_genre_fem.plot(kind='bar',y=['females_allages_avg_vote'])
```
